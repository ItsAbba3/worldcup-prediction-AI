```python
import json
from predict import predict_upcoming_match
from model import train

model = train()
matches = json.load(open('fifa_world_cup_2026_group_matches.json', encoding='utf-8'))['matches']

results = []
for m in matches:
    try:
        pred = predict_upcoming_match(
            match_name=f"{m['home']} vs {m['away']}",
            venue=m['venue'],
        )
        results.append({
            "match_id": m['match_id'],
            "home": m['home'],
            "away": m['away'],
            "moneyline": {
                "home": pred.prob_home_win,
                "draw": pred.prob_draw,
                "away": pred.prob_away_win,
            },
            "xg": {
                "home": round(pred.xg_home, 3),
                "away": round(pred.xg_away, 3),
            },
            "top_scores": pred.top_scorelines[:5],
            "btts": pred.prob_btts if hasattr(pred, 'prob_btts') else None,
            "totals": pred.totals if hasattr(pred, 'totals') else {},
        })
        print(f"OK: {m['home']} vs {m['away']}")
    except Exception as e:
        print(f"ERR {m['home']} vs {m['away']}: {e}")

with open('wc26_matches.json', 'w', encoding='utf-8') as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

print(f"\nDone: {len(results)} matches saved to wc26_matches.json")
```

```python
import json
from predict import predict_upcoming_match

matches = json.load(open('fifa_world_cup_2026_group_matches.json', encoding='utf-8'))['matches']

results = []
for m in matches:
    try:
        pred = predict_upcoming_match(
            match_name=f"{m['home']} vs {m['away']}",
            venue=m['venue'],
        )
        results.append({
            "match_id": m['match_id'],
            "home": m['home'],
            "away": m['away'],
            "moneyline": {
                "home": pred.prob_home_win,
                "draw": pred.prob_draw,
                "away": pred.prob_away_win,
            },
            "xg": {
                "home": round(pred.xg_home, 3),
                "away": round(pred.xg_away, 3),
            },
            "top_scores": [{"score": s["score"], "p": s["probability"]} for s in pred.top_scorelines[:5]],
            "totals": getattr(pred, 'totals', {}),
            "btts": getattr(pred, 'prob_btts', None),
        })
        print(f"OK: {m['home']} vs {m['away']}")
    except Exception as e:
        print(f"ERR {m['home']} vs {m['away']}: {e}")

with open('wc26_matches.json', 'w', encoding='utf-8') as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

print(f"Done: {len(results)} matches")
```
