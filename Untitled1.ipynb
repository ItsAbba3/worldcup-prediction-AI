```python
futures = results["futures"]
sorted_teams = sorted(futures.items(), key=lambda x: x[1]["champion"], reverse=True)

for team, odds in sorted_teams:
    print(f"{team:25s}  قهرمانی: {odds['champion']*100:.1f}%  |  فینال: {odds.get('final',0)*100:.1f}%  |  صعود: {odds.get('advance',0)*100:.1f}%")
```
